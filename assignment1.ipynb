```python
sample_string =  "Hello My name is si    emanYMoLLeH"
sample_string2 =  "Hello My name is si    edsfmanYMoLLeH"


def check_palindrome(strong):
    newstring = ""
    #should return true
    for i in strong:
        if i != ' ':
            newstring += i.lower()


    length = len(newstring)

    for i in range(length):
        if newstring[i] == newstring[length-1-i]:
            continue
        else:
            return False


    return True


print(check_palindrome(sample_string2))
print(check_palindrome(sample_string))
```

**Q3**

```python
def pair_diff(no_list : list ,length :int ,x : int) :
    for i in range(length):
        for j in range(i+1,length):
            if abs(no_list[i] - no_list[j]) == x:
                return no_list[i],no_list[j]


def pair_diff_fast(no_list: list, x: int):
    # Convert the list to a set for instant O(1) lookups
    seen_numbers = set(no_list)

    for num in no_list:
        # Check if the needed pair exists in our set
        if (num + x) in seen_numbers:
            return num, num + x
```

```python
daily_returns = np.random.normal(loc = 0.0008,scale = 0.015,size = 252)

price_series =[100]

for i in range(252):
    price_series.append(price_series[i]*(1+daily_returns[i]))
print("start day price " + str(price_series[0]))
print("last day price " + str(price_series[252]))


max = float('-inf')
for i in price_series:
    if max < i:
        max = i
print("max price " + str(max))

#find peak and the corresponding trough then calculate the drawdown


peak = price_series[0] - 1

curr_index = 0
maxdrawdown = 0
running_peak = price_series[0]

for current_price in price_series:

    if current_price > running_peak:
        running_peak = current_price

    current_drawdown = ((running_peak - current_price) / running_peak) * 100

    if current_drawdown > maxdrawdown:
        maxdrawdown = current_drawdown

print("max drawdown % = " + str(maxdrawdown) + "%")

fig , ax = plt.subplots()
x =[i for i in range(1,254,1)]

ax.set_xlabel("Days")
ax.set_ylabel("Daily Prices")
ax.set_title("Stock Prices")


ax.plot(x,price_series,linewidth = 1.2)

plt.show()
```

```python
np.random.seed(42)
returns = np.random.multivariate_normal(
mean=[0.0005, 0.0008, 0.0003],
cov=[[0.0004, 0.0001, 0.00005],
[0.0001, 0.0006, 0.00008],
[0.00005, 0.00008, 0.0003]],
size=200
)
prices = 100 * np.cumprod(1 + returns, axis=0)



col_name = ["Asset A","Asset B","Asset C"]

df = pd.DataFrame(prices,columns=col_name)

asset_a = df["Asset A"]
asset_b = df["Asset B"]
asset_c = df["Asset C"]

portfolio = (asset_a+asset_b+asset_c)/3

days = [i for i in range(1,201)]


returns_data = df.pct_change()

returns_data = df.pct_change().dropna()

centered_returns = returns_data - returns_data.mean()

covariation_matrix = (centered_returns.T @ centered_returns)/(len(centered_returns) - 1)

variances = np.diag(covariation_matrix)

std_devs = np.sqrt(variances)

std_grid = np.outer(std_devs, std_devs)

correlation_matrix = covariation_matrix / std_grid

print(correlation_matrix)

fig ,ax = plt.subplots(figsize = (12,6))
ax.set_xlabel("Days")
ax.set_ylabel("Asset Prices")

ax.plot(days,asset_a,label = "Asset A",linewidth = 2)
ax.plot(days,asset_b,label = "Asset B",linewidth = 2)
ax.plot(days,asset_c,label = "Asset C",linewidth = 2)
ax.plot(days,portfolio,label = "Portfolio(1/3)",linewidth = 2)

plt.legend()
plt.show()

print(df.head())
```
